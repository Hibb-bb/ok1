# CIFAR-10 reconstruction grid (`vis.py`)

Runs **`vis.py`** via subprocess (`cwd = ok1/`) so `torchvision` data paths (`./data/`) and **`--device cuda`** geomstats hints behave like the CLI.

We pass **`--no-show`** so the notebook does not block on `plt.show()`; the figure is written to **`--out`** and displayed below.

### Setup
- Kernel: **`ok1/.venv`**.
- `%matplotlib inline` is optional (subprocess manages its own Matplotlib).

```python
%matplotlib inline
```

```python
from __future__ import annotations

import subprocess
import sys
from pathlib import Path


def find_repo_root() -> Path:
    here = Path.cwd().resolve()
    candidates = [here, *here.parents[:4]]
    for d in candidates:
        if (d / "vis.py").is_file():
            return d
    raise RuntimeError("Could not find vis.py.")


REPO_ROOT = find_repo_root()
print("REPO_ROOT =", REPO_ROOT)


def run_vis(*argv: str) -> None:
    script = REPO_ROOT / "vis.py"
    cmd = [sys.executable, str(script), *argv]
    print("$", " ".join(cmd))
    subprocess.run(cmd, cwd=str(REPO_ROOT), check=True)
```

### Generate grid (moderate `--N`; first run downloads CIFAR-10)

```python
OUT_PNG = REPO_ROOT / "outputs" / "vis" / "notebook_recon_demo.png"
OUT_PNG.parent.mkdir(parents=True, exist_ok=True)

run_vis(
    "--N",
    "40",
    "--sigma",
    "0.2",
    "0.45",
    "--no-pca",
    "--beta",
    "10",
    "--no-show",
    "--out",
    str(OUT_PNG),
)
```

### Show saved PNG

```python
from IPython.display import Image, display

if OUT_PNG.is_file():
    display(Image(filename=str(OUT_PNG)))
else:
    raise FileNotFoundError(OUT_PNG)
```
